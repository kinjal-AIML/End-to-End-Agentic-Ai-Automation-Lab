# **Async Programming**

### **Synchronous Programming**

```python
import time

def brew_coffee():
    print("Brewing Coffee")
    time.sleep(3)
    print("Coffee is ready...")
    

def tost_bagel():
    print("Toasting Bagel")
    time.sleep(2)
    print("Bagel is Ready..")
    
    
def main():
    start = time.time()
    
    coffee = brew_coffee()
    bagel = tost_bagel()
    
    end = time.time()
    
    print(f"Time: {end-start:.2f} minutes")
```

```python
main()
```

## **Async Programming**

```python
import asyncio

import time

async def brew_coffee():
    print("Brewing Coffee")
    await asyncio.sleep(3)
    print("Coffee is ready...")
    

async def tost_bagel():
    print("Toasting Bagel")
    await asyncio.sleep(2)
    print("Bagel is Ready..")
    
    
async def main():
    start = time.time()
    
    coffee = brew_coffee()
    bagel = tost_bagel()
    
    
    result = await asyncio.gather(coffee, bagel)
    end = time.time()
    
    print(f"Time: {end-start:.2f} minutes")

await main()
```

## **Rule-1**
- Always call async function with `await`.
