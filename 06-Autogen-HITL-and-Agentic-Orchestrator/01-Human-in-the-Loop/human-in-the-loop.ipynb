## **UserProxy Agent**

![alt text](hitl-userProxy.png "Human in The loop User proxy")

The bold arrows indicates the flow of control during a team’s run: when the team calls the `UserProxyAgent`, it transfers the control to the application/user, and waits for the feedback; once the feedback is provided, the control is transferred back to the team and the team continues its execution.

## **User Proxy Agent**

```python
from autogen_agentchat.agents import UserProxyAgent
from autogen_agentchat.teams import RoundRobinGroupChat

user_proxy_agent = UserProxyAgent(
    name="User_Proxy",
    description="You are a user proxy agent.",
    input_func=input
)
```

```python
from autogen_agentchat.conditions import TextMentionTermination

termination_condition = TextMentionTermination(text="APPROVE")
```

```python
team = RoundRobinGroupChat(
    participants=[assistant, user_proxy_agent],
    termination_condition=termination_condition,
    max_turns=10
    
)
```

```python
from autogen_agentchat.messages import TextMessage
from autogen_agentchat.ui import Console
```

```python
async def run_team():
    task = TextMessage(content="Write a poem about the Bangladesh.", source="user")
    
    result = await team.run(task=task)
    
    for each_agent_msg in result.messages:
        print(f"{each_agent_msg.source} :-> {each_agent_msg.content}")
    
await run_team()
```
