```python
from langchain_mcp_adapters.client import MultiServerMCPClient

clint = MultiServerMCPClient(
    {
        "local_server": {
            "transport": "stdio",
            "command": "python",
            "args": ["mcp_server.py"]
        }
    }
)
```

## **Online MCP Server**

```python
clint1 = MultiServerMCPClient(
    {
  
    "time": {
      "transport": "stdio",
      "command": "uvx",
      "args": [
        "mcp-server-time",
        "--local-timezone=America/New_York"
      ]
    }
  
    }
)
```

```python
tools = await clint.get_tools()
tools
```

```python
resources = await clint.get_resources("local_server")
blob = resources[0]

print(blob.model_dump())
```

```python
prompt_blocks = await clint.get_prompt("local_server", "prompt")
system_prompt = prompt_blocks[0].content

print(system_prompt)
```

```python
import os
from dotenv import load_dotenv
load_dotenv()

from langchain.chat_models import init_chat_model
from langchain.agents import create_agent

model = init_chat_model("gpt-4.1-mini")

agent = create_agent(
    model=model,
    tools=tools,
    system_prompt=system_prompt
)
```

```python
from langchain.messages import HumanMessage

response = await agent.ainvoke(
    {
        "messages": [
            HumanMessage(content="Explain what LoRA is based on the repo")
        ]
    }
)

print(response["messages"][-1].content)
```
