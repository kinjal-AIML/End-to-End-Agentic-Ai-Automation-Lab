```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python

os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")

from langchain.chat_models import init_chat_model

model = init_chat_model("gpt-4.1-mini")

response = model.invoke("What is the capital of the moon")
response
```

```python
response.response_metadata
```

```python
from pprint import pprint

pprint(response.response_metadata)
```

## **Langchain agent Creations**

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage

agent = create_agent(model=model)

response = agent.invoke(
    {
        "messages": [
            HumanMessage(content="WHat is the capital of the moon?")
        ]
    }
)

pprint(response["messages"])
```

```python
for msg in response["messages"]:
    pprint(msg.content)
```

## **Striming output**

```python
for token, metadata in agent.stream(
    {
        "messages": [
            HumanMessage(content="WHat is the capital of the moon?")
        ]
    },
    stream_mode="messages"
):
    if token.content:
        print(token.content, end="", flush=True)
```

## **Agent with system prompts**

```python
system_msg = "You're a science fictions writer, create a capital city for the user request."

sicFicAgent = create_agent(
    system_prompt=system_msg,
    model=model
)

response = sicFicAgent.invoke(
    {
        "messages": [HumanMessage(content="What is the capital of the moon?")]
    }
)
```

## **Structure output from the agent**

```python
from pydantic import BaseModel

class CapitalInfos(BaseModel):
    name: str
    locations: str
    vibe: str
    economy: str
```

```python
own_agent = create_agent(
    system_prompt=system_msg,
    model=model,
    response_format=CapitalInfos
)
```

```python
response = own_agent.invoke(
    {
        "messages": [HumanMessage(content="What is the capital of the moon?")]
    }
)
```
