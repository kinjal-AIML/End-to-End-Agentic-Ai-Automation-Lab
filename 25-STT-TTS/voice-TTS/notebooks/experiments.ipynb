## **Voice ai `API` Testing**

```python
from dotenv import load_dotenv
import os
load_dotenv()

voice_api_key = os.getenv("VOICE_API_KEY")
```

```python
BASE_URL = "https://dev.voice.ai"
```

## **Generate Speech**

```python
import requests

# Using default voice (voice_id is optional)
response = requests.post(
    'https://dev.voice.ai/api/v1/tts/speech',
    headers={'Authorization': f'Bearer {voice_api_key}', 'Content-Type': 'application/json'},
    json={
        'text': 'Hi, i"m Al Amin. How can i help you.',
        'model': 'voiceai-tts-v1-latest',  # Optional, defaults to voiceai-tts-v1-latest
        'language': 'en'  # Optional, defaults to 'en'
    }
)

# Or with a custom voice_id:
# json={'voice_id': 'your-voice-id-here', 'text': 'Hello! This is a test of the Voice.ai TTS API.', 'model': 'voiceai-tts-v1-latest', 'language': 'en'}

with open('output.mp3', 'wb') as f:
        f.write(response.content)
```

## **Test different voice**

```python
import json
import requests
from pathlib import Path

voice_api_key = voice_api_key

API_URL = "https://dev.voice.ai/api/v1/tts/speech"

# output folder
output_dir = Path("audio_outputs")
output_dir.mkdir(exist_ok=True)

# load json
with open("tts_data.json") as f:
    data = json.load(f)

for item in data["stt_test_data"]:
    text = item["text"]
    file_name = f"{item['id']:03d}_{item['category']}.mp3"
    file_path = output_dir / file_name

    response = requests.post(
        API_URL,
        headers={
            "Authorization": f"Bearer {voice_api_key}",
            "Content-Type": "application/json"
        },
        json={
            "text": text,
            "model": "voiceai-tts-v1-latest",
            "language": "en"
        }
    )

    if response.status_code == 200:
        with open(file_path, "wb") as f:
            f.write(response.content)
        print(f"Saved: {file_path}")
    else:
        print(f"Failed for {file_name}: {response.text}")
```
