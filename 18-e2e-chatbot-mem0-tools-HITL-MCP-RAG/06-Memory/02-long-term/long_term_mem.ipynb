
# **Long Term memory for Agent**
### There are 3 types of long-term memory
- Procedural memory
- Semantic memory
- Episodic memory

```python
from langgraph.store.memory import InMemoryStore

store = InMemoryStore()

namespace = ("user", "u1")
```

## **Creating Memory**

```python
store.put(namespace, "1", {"data": "user like pizza."})
store.put(namespace, "2", {"data": "user have a Honda bike."})
```

```python
# another namespace
namespace2 = ("user", "u2")

# Add memories
store.put(namespace2, "1", {"data": "User likes pasta"})
store.put(namespace2, "2", {"data": "User prefers grid style navigation"})
```

## **Retrieving Memory**
- fetching `user1` key_1 memory

```python
store.get(namespace=namespace, key="1")
```

## **Retrieving all memories**

```python
items = store.search(namespace)

for item in items:
    print(item.value)
```

## **Semantic Search**

```python
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv

load_dotenv()

embedding_model = OpenAIEmbeddings(model='text-embedding-3-small')
```

```python
store = InMemoryStore(
    index={
        "embed": embedding_model,
        "dims": 1536
    }
)

namespace = ("users", "alamin")
```

```python
store.put(namespace, "1", {"data": "User prefers concise answers over long explanations"})
store.put(namespace, "2", {"data": "User likes examples in Python"})
store.put(namespace, "3", {"data": "User usually works late at night"})
store.put(namespace, "4", {"data": "User prefers dark mode in applications"})
store.put(namespace, "5", {"data": "User is learning agentic ai, llm from scratch."})
store.put(namespace, "6", {"data": "User dislikes overly theoretical explanations"})
store.put(namespace, "7", {"data": "User prefers step-by-step reasoning"})
store.put(namespace, "8", {"data": "User is based in Bangladesh"})
store.put(namespace, "9", {"data": "User likes real-world analogies"})
store.put(namespace, "10", {"data": "User prefers bullet points over paragraphs"})
```

## How the search is performing.
- this query goes to the `namespace` and search the query related data.
- before in vanilla search we take all mem but now we take only what we need exactly?

```python
items = store.search(namespace, query="what is the user currently learning", limit=1)

for item in items:
    print(item.value)
```

```python
items = store.search(namespace, query="what is the user's preferences?", limit=3)

for item in items:
    print(item.value)
```
