
# **Long Term memory for Agent**
### There are 3 types of long-term memory
- Procedural memory
- Semantic memory
- Episodic memory

```python
from langgraph.store.memory import InMemoryStore

store = InMemoryStore()

namespace = ("user", "u1")
```

## **Creating Memory**

```python
store.put(namespace, "1", {"data": "user like pizza."})
store.put(namespace, "2", {"data": "user have a Honda bike."})
```

```python
# another namespace
namespace2 = ("user", "u2")

# Add memories
store.put(namespace2, "1", {"data": "User likes pasta"})
store.put(namespace2, "2", {"data": "User prefers grid style navigation"})
```

## **Retrieving Memory**
- fetching `user1` key_1 memory

```python
store.get(namespace=namespace, key="1")
```

## **Retrieving all memories**

```python
items = store.search(namespace)

for item in items:
    print(item.value)
```
