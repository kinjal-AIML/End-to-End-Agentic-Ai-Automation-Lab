
# **Long Term memory for Agent**
### There are 3 types of long-term memory
- Procedural memory
- Semantic memory
- Episodic memory

```python
from langgraph.store.memory import InMemoryStore

store = InMemoryStore()

namespace = ("user", "u1")
```

## **Creating Memory**

```python
store.put(namespace, "1", {"data": "user like pizza."})
store.put(namespace, "2", {"data": "user have a Honda bike."})
```
